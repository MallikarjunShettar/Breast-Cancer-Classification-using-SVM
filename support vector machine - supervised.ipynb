```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import metrics
```

```python
y_pred = cls.predict(X_test)
```

```python
recall_score(Y_test,y_pred)
```

## Data Standardization

```python
#Standardization refers to shifting the distribution of each attribute to have a mean of zero
#and a standard deviation of one (unit variance). It is useful to standardize attributes for a model. 
#Standardization of datasets is a common requirement for many machine learning estimators implemented in scikit-learn;
#they might behave badly if the individual features do not more or less look like standard normally distributed data.
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.8,random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
```

```python
clf.fit(X_train,Y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
accuracy_score(Y_test,y_pred)
```

```python
precision_score(Y_test,y_pred)
```

```python
recall_score(Y_test,y_pred)
```

```python
f1_score(Y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(Y_test,y_pred)
```

```python
# Kernel=RBF(Radial Basis Function)

clf1 = SVC(kernel='rbf')
clf1.fit(X_train,Y_train)
y_pred1 = clf1.predict(X_test)
print('Accuracy Score:')
print(metrics.accuracy_score(Y_test,y_pred1))
```

```python
# Kernel = Ploynomial

svc = SVC(kernel='poly')
svc.fit(X_train,Y_train)
y_pred2=svc.predict(X_test)
print('Accuracy Score:')
print(metrics.accuracy_score(Y_test,y_pred2))
```
